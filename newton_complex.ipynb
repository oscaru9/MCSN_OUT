```julia
function fx(x, a)
    return x^4 + a
    
end
```

```julia
function dfx(x)
    return 4*x^3
    
end
```

```julia
n=1000
function newton(x0, a)
    x=zeros(n+1)
    x[1]=x0
    for i=1:n
        x[i+1]=x[i]-fx(x[i],a)/dfx(x[i])
        
        
    end
    return x
end
```

```julia
y=newton(1, 4)

y1=newton(1.1, 4)
```

```julia
using Plots

plot(x, y[1:100], label="x0=1", xlabel="iteration", ylabel="x", title="Newton method")
plot!(x, y1[1:100], label="x0=1.1", xlabel="iteration", ylabel="x", title="Newton method")
```

```julia
Iniciales=[x + y*im for x in -1:0.001:1, y in -1:0.001:1];
```

```julia
Iniciales
```

```julia
f(x)= x-(x^4 + 4)/(4*x^3)
```

```julia
iterator = deepcopy(Iniciales)
[iterator .= f.(iterator) for i in 1:100] 
```

```julia
g(x)= x-(x^4 - 4)/(4*x^3)
```

```julia
iterator2 = deepcopy(Iniciales)
[iterator2 .= g.(iterator2) for i in 1:100]
```

```julia
function raices(A)
    raices= Set{Any}()
    for i in 1:size(A)[1]
        for j in 1:size(A)[2]
            try 
                x=f(A[i,j])
                push!(raices, x)
            catch
                
            end
        end
        return collect(raices)
    end
end
```

```julia
raices(iterator)
```

```julia
function comparacion(A, r, tol=1e-3)
    
    B=zeros(size(A)[1], size(A)[2])
    for j in 1:size(A)[1]*size(A)[2]
        if abs(A[j]-r[1])<tol
            B[j]= 1
        elseif abs(A[j]-r[2])<tol
            B[j]= 2
        elseif abs(A[j]+r[3])<tol
            B[j]= 3
        elseif abs(A[j]-r[4])<tol
            B[j]=4
        else
            B[j]=0
        end
    end
    return B
end
    
        
```

```julia
raiz=[1+ 1im, -1+1im, -1-1im, 1-1im]
R=comparacion(iterator, raiz)
```

```julia
heatmap(R, title="Fractal de Newton", c=:viridis)
```

```julia
raiz2=[sqrt(2), sqrt(2)*im, -sqrt(2), -sqrt(2)*im]
R2=comparacion(iterator2, raiz2)
```

```julia
heatmap(R2, title="Fractal de Newton", c=:viridis, size=(1200, 800))
savefig("fractal_newton.png")
```
