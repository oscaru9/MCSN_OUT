```julia
function newton(x0, f, delta, tol=1e-5, max_itt=100)
    x=x0
    for i in 1:max_itt
        derivate=(f(x+delta)-f(x))/delta
        x= x -f(x)/derivate
        if abs(f(x))<tol
            return x, i
        end
    end
    
end
```

```julia
f(x)=x^2-16
```

```julia
y=[]
x=[]
for i in 1:0.1:20
    push!(y,newton(3, f, 5/2^(i+2))[2])
    push!(x, 5/2^(i+2))
end
```

```julia
using Plots
using LaTeXStrings

plot(x,y, label="x0=3", xlabel=L"\Delta x", ylabel=L"N(\Delta x)", title="Newton method")
```
