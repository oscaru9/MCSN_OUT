Queremos usar el método de Newton-Raphson para encontrar las raíces de la función A=x^2sec^2(x)

```julia
function Ecuacion_trascendental(x, A)
    return x*tan(x)-sqrt(A-x^2)
end
```

```julia
function newton_a(x0, A, tol=1e-6, max_iter=100)
    x = x0
    
    for i in 1:max_iter
        f= Ecuacion_trascendental(x, A)
        df = tan(x) + x*(sec(x))^2 + x/sqrt(A-x^2)
        x = x - f/df 
        if abs(f) < tol
            return x
        end
    end
    return x
    
end
```

```julia
function raices(A)
    raices = Set{Any}()
    for i in -10:10
        try
        x = newton_a(i, A)
        push!(raices, x)
    
        catch
        end
    end
    return collect(raices)
end
```

```julia
root_=raices(10)
```

```julia
using Plots
using LaTeXStrings
function xy()
    x = -3:0.01:3
    
    y = [Ecuacion_trascendental(i, 10) for i in x]
    return x, y
end

plot(xy()[1], xy()[2], label=L"f(x)=x \tan(x) - \sqrt{10 - x^2}", xlabel="x", ylabel="f(x)", title="Ecuación trascendental")
ylims!(-10, 10)
plot!(root_, [0 for i in root_], seriestype=:scatter, label="Raices")
```
